### Clean Text Columns

```python
#Selecting text columns
text_cols=df.select_dtypes(include=["object","string"]).columns

#Cleaning text columns by stripping leading and trailing spaces
for col in text_cols:
    df[col]=df[col].astype(str).str.strip()


#Code Explained:
#df.select_dtypes(include=["object","string"])
"""selects all columns in the DataFrame that have data types of either "object" or "string"."""

#.columns
"""retrieves the column names of the selected columns."""

#df[col].astype(str)
"""converts the values in the column to string data type."""

#.str.strip()
"""removes leading and trailing whitespace characters from the string values in the column."""

#.str.lower()
"""converts the string values to lowercase."""

"""This code is used to clean text columns in a DataFrame by removing leading and trailing spaces and
converting the text to lowercase. It ensures that the text data is standardized and consistent for further
analysis or processing."""
```

### Handle Missing Values

```python
#Checking missing value percentages for each column
missing_percent=round(df.isnull().mean()*100,2)
print(missing_percent.sort_values(ascending=False))
```

```python
#Filling numeric columns
num_cols=df.select_dtypes(include=["number"]).columns

for col in num_cols:
    df[col]=df[col].fillna(df[col].median())

#Filling object columns
obj_cols = df.select_dtypes(include=["object", "string"]).columns

for col in obj_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

#Filling datetime columns
datetime_cols=df.select_dtypes(include=["datetime64[ns]"]).columns

for col in datetime_cols:
    df[col]=df[col].fillna(df[col].mode()[0])

"""Although the dataset doesn't contain any datetime columns at this moment, 
this code is included to handle any future datetime columns that may be added to the dataset."""

#.mode()[0]
"""calculates the mode (most frequent value) of the column
and selects the first mode in case there are multiple modes."""
```

### Clean Latitude & Longitude

```python
#Removing impossible coordinates
df.loc[
    (df["latitude"] < -90) |
    (df["latitude"] > 90),
    "latitude"
] = np.nan

df.loc[
    (df["longitude"] < -180) |
    (df["longitude"] > 180),
    "longitude"
]=np.nan


#df.loc[row_condition, column_name] 
"""This is the standard method to select specific rows and columns"""

#The Condition (df["latitude"] < -90) | (df["latitude"] > 90)
"""This looks for any row where the latitude is less than -90 OR (represented by the | symbol) greater than 90.
These are physically impossible latitudes."""

#=np.nan 
"""For every broken row found, it overwrites the bad latitude with a blank/missing value (NaN)"""

"""Because the Earth is a sphere, coordinates have strict physical boundaries:
 Latitude must be between -90° and 90°.
 Longitude must be between -180° and 180°.

The .loc property looks for rows where the latitude is less than -90 or (|) greater than 90.
The result: If it finds an impossible value (like a latitude of 105.2),
it zeroes it out by replacing it with np.nan (Not a Number/missing value)."""
```

```python
#Fill
df["latitude"]=df["latitude"].fillna(df["latitude"].median())
df["longitude"]=df["longitude"].fillna(df["longitude"].median())

"""Once the bad data has been wiped out and turned into NaN,
the next two lines attempt to patch those null values with median."""
```

### Clean Property Loss Column

```python
"""The Property Loss (USD) column contains values like:
    #60.75.0
    44839.49
    NaN"""

#Convert safely:
df["property_loss_usd"]=(df["property_loss_usd"].astype(str).str.replace(r"[^0-9.]","",regex=True))
```

```python
#Fixing multiple decimal points:
def fix_amount(x):
    try:
        parts=x.split(".")
        
        if len(parts) > 2:
            x=parts[0] + "." + "".join(parts[1:])
            
        return float(x)
    
    except:
        return np.nan

df["property_loss_usd"]=df["property_loss_usd"].apply(fix_amount)

"""This function:
    Looks for values with multiple decimal points.
    Keeps only the first decimal point.
    Converts the result into a numeric value (float).
    Replaces unfixable values with NaN."""
```

```python
#Filling missing:
df["property_loss_usd"]=df["property_loss_usd"].fillna(df["property_loss_usd"].median())
```
