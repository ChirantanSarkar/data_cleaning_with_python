### Step 3: Remove Duplicate Rows

```python
#Removing duplicate rows from the dataset
df=df.drop_duplicates()
print(df.shape)
```

### Step 4: Standardize Column Names

```python
#Fixing column format
df.columns=(df.columns.str.lower().str.strip().str.replace(" ", "_"))
print(df.columns)
```

### Step 5: Clean Text Columns

```python
#Selecting text columns
text_cols=df.select_dtypes(include=["object","string"]).columns

#Cleaning text columns by stripping leading and trailing spaces
for col in text_cols:
    df[col]=df[col].astype(str).str.strip()
```

### Step 6: Handle Missing Values

```python
#Checking missing value percentages for each column
missing_percent=round(df.isnull().mean()*100,2)
print(missing_percent.sort_values(ascending=False))
```

```python
#Filling numeric columns
num_cols=df.select_dtypes(include=["number"]).columns

for col in num_cols:
    df[col]=df[col].fillna(df[col].median())

#Filling datetime columns
datetime_cols=df.select_dtypes(include=["datetime64[ns]"]).columns

for col in datetime_cols:
    df[col]=df[col].fillna(df[col].mode()[0])

#Filling object columns
obj_cols=df.select_dtypes(include="object").columns

for col in obj_cols:
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
#Checking missing values in the dataset after cleaning
df.isna().sum()
```

### Step 7: Fix Datetime Column

```python
#Current datatype of incident_datetime column
df["incident_datetime"].dtype
```

```python
#Convert column type to datetime
df["incident_datetime"]=pd.to_datetime(df["incident_datetime"],errors="coerce")

#errors="coerce" will convert any invalid date formats to NaT (Not a Time) instead of raising an error. 
#This is useful when we have a column with mixed date formats or some invalid entries that cannot be converted to datetime.
```

```python
#Checking the datatype of incident_datetime column after conversion
df["incident_datetime"].dtype

#dtype("<M8[us]") means that the column is of type datetime64 with microsecond precision.
#The "<" indicates that the data is stored in little-endian byte order, which is the default for most systems.
```

```python
#Checking if there are any missing values in the dataset after cleaning
df.isna().sum()

#There are 380 missing values in the incident_datetime column after cleaning.
#The datatype of these values cannot be changed.
#Either we can drop these rows or fill them with a specific value.
```

### Step 8: Clean Latitude & Longitude

```python
#Removing impossible coordinates
df.loc[
    (df["latitude"] < -90) |
    (df["latitude"] > 90),
    "latitude"
] = np.nan

df.loc[
    (df["longitude"] < -180) |
    (df["longitude"] > 180),
    "longitude"
]=np.nan


#Because the Earth is a sphere, coordinates have strict physical boundaries:
 #Latitude must be between -90° and 90°.
 #Longitude must be between -180° and 180°.

#The .loc property looks for rows where the latitude is less than -90 or (|) greater than 90.
#The result: If it finds an impossible value (like a latitude of 105.2), it zeroes it out by replacing it with np.nan (Not a Number/missing value).
```

```python
#Fill
df["latitude"]=df["latitude"].fillna(df["latitude"].median())
df["longitude"]=df["longitude"].fillna(df["longitude"].median())

#Once the bad data has been wiped out and turned into NaN, the next two lines attempt to patch those null values with median.
```

### Step 9: Clean Property Loss Column

```python
#The Property Loss (USD) column contains values like:
    #60.75.0
    #44839.49
    #NaN

#Convert safely:
df["property_loss_usd"]=(df["property_loss_usd"].astype(str).str.replace(r"[^0-9.]","",regex=True))
```

```python
#Fixing multiple decimal points:
def fix_amount(x):
    try:
        parts=x.split(".")
        
        if len(parts) > 2:
            x=parts[0] + "." + "".join(parts[1:])
            
        return float(x)
    
    except:
        return np.nan

df["property_loss_usd"]=df["property_loss_usd"].apply(fix_amount)

#This function:
    #Looks for values with multiple decimal points.
    #Keeps only the first decimal point.
    #Converts the result into a numeric value (float).
    #Replaces unfixable values with NaN.
```

```python
#Filling missing:
df["property_loss_usd"]=df["property_loss_usd"].fillna(df["property_loss_usd"].median())
```

### Step 10: Standardize Severity


```python
#Current values include:
 #1,2,3,4
 #Low
 #low
 #MEDIUM
 #Med
 #High
 ##Critical
 ##Crit


#Create mapping:
severity_map = {
    "1":"Low",
    "low":"Low",
    "Low":"Low",

    "2":"Medium",
    "Med":"Medium",
    "MEDIUM":"Medium",
    "Medium":"Medium",

    "3":"High",
    "high":"High",
    "High":"High",

    "4":"Critical",
    "Crit":"Critical",
    "CRITICAL":"Critical",
    "Critical":"Critical"
}

#Replacing values 
df["severity"]=df["severity"].replace(severity_map)
```

```python
#Checking value counts 
df["severity"].value_counts()
```

### Step 11: Standardize Case Status

```python
#Current values:
 # Open
 # OPEN
 # open
 # Closed
 # CLOSED
 # Resolved
 # Pendng 
 # Pending
 # Investgation
 # Under Investigation

#Mapping:
status_map = {
    "OPEN":"Open",
    "open":"Open",
    "Open":"Open",

    "Closed":"Closed",
    "closed":"Closed",
    "CLOSED":"Closed",

    "Resolved":"Resolved",

    "Pendng":"Pending",
    "Pending":"Pending",

    "Investgation":"Under Investigation",
    "under investigation":"Under Investigation",
    "Under Investigation":"Under Investigation"
}

#Replacing values
df["case_status"]=df["case_status"].replace(status_map)
```

```python
#Checking unique values:
df["case_status"].value_counts()
```

### Step 12: Standardize Reported Online

```python
#Current values:
 # True
 # False
 # YES
 # NO
 # 1
 # 0
 # Yes
 # No

#Converting to binary values:
online_map = {
    "YES":1,
    "Yes":1,
    "yes":1,
    "True":1,
    "1":1,

    "NO":0,
    "No":0,
    "no":0,
    "False":0,
    "0":0
}

#Replacing values
df["reported_online"]=(df["reported_online"].replace(online_map).astype(int))
```

```python
#Checking unique values:
df["reported_online"].value_counts()
```

### Step 13: Clean Age Columns

```python
age_cols=["suspect_age","victim_age"]

for col in age_cols:
    df.loc[(df[col] < 0) | (df[col] > 120), col]=np.nan
    df[col]=df[col].fillna(df[col].median())

#This code performs two data-cleaning tasks on the age columns:
 #Identifies impossible ages and marks them as missing (NaN)
 #Replaces all missing ages with the median age
```

### Step 14: Standardize Crime Types

```python
#The dataset contains many misspellings:
# asslt
# Burglry
# Homocide
# Arsen
# Domestc Violence
# B&E
# CYBER CRIME

#Create mapping:
crime_map = {
    "asslt":"Assault",
    "Burglry":"Burglary",
    "burglary":"Burglary",
    "Homocide":"Homicide",
    "Arsen":"Arson",
    "Domestc Violence":"Domestic Violence",
    "B&E":"Breaking & Entering",
    "THEFT":"Theft",
    "CYBER CRIME":"Cyber Crime",
    "Drug Offense":"Narcotics"
}

#Replacing values
df["crime_type"]=df["crime_type"].replace(crime_map)
```

```python
#Checking unique values:
df["crime_type"].value_counts()
```

### Step 15: Standardize Gender Values

```python
#Mapping
gender_map={
    "M":"Male",
    "m":"Male",
    "Male":"Male",
    "MALE":"Male",
    "male":"Male",

    "F":"Female",
    "f":"Female",
    "FEMALE":"Female",
    "Female":"Female",
    "female":"Female"
}

#Replacing values
df["suspect_gender"]=df["suspect_gender"].replace(gender_map)
df["victim_gender"]=df["victim_gender"].replace(gender_map)
```
